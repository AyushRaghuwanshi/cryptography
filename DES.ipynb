```python
def keygenerator(text):
    #56 key generation
    key_64 = textTobit(text)
    key_56 = []
    for i in PC_1:
        key_56.append(key_64[i-1])
    l0 = key_56[0:28]
    r0 = key_56[28:]
    l_1_16 = []
    r_1_16 = []
    for i in range(1,17):
        if(i in [1,2,9,16]):
            temp_l = l0[0]
            temp_r = r0[0]
            for j in range(27):
                l0[j] = l0[j+1]
                r0[j] = r0[j+1]
            l0[27] = temp_l
            r0[27] = temp_r
            l_1_16.append(l0.copy())
            r_1_16.append(r0.copy())
        else:
            temp_l_1 = l0[0]
            temp_l_2 = l0[1]
            temp_r_1 = r0[0]
            temp_r_2 = r0[1]
            for j in range(26):
                l0[j] = l0[j+2]
                r0[j] = r0[j+2]
            l0[27] = temp_l_2
            r0[27] = temp_r_2
            l0[26] = temp_l_1
            r0[26] = temp_r_1
            l_1_16.append(l0.copy())
            r_1_16.append(r0.copy())
    key_full_56 = []
    for i in range(16):
        key_full_56.append(l_1_16[i]+r_1_16[i])
    key_48 = []
    for i in range(16):
        temp = []
        for j in PC_2:
            temp.append(key_full_56[i][j-1])
        key_48.append(temp.copy())
    return key_48

def addBit(text):
    extended_text = []
    for i in E_bit_s_t:
        extended_text.append(text[i-1])
    return extended_text


def textTobit(text):
    text = text[:16]
    return "{0:064b}".format(int(text, 16))


    
def permutateIP(text):
    new_text = ''
    for i in IP:
        new_text += str(text[i-1])
    return new_text

def xoroperation(text1, text2):
    new_text = ''
    for i in range(len(text1)):
        if(text1[i] == text2[i]):
            new_text += '0'
        else:
            new_text += '1'
    return new_text

def convert32_48(text):
    new_text = ''
    for i in E_bit_s_t:
        new_text += text[i-1]
    return new_text

def convert48_32(text):
    new_text = ''
    for i in range(8):
        temp = text[i*6:i*6+6]
        row = temp[0] + temp[5]
        row = int(row, 2)
        col = temp[1:5]
        col = int(col, 2)
        val = s_box[i][row][col]
        val = "{0:04b}".format(int(str(val), 10))
        new_text += val
    return new_text
        
def p_each_round(text):
    new_text = ''
    for i in P:
        new_text += text[i-1]
    return new_text

def p_final_round(text):
    new_text = ''
    for i in FI:
        new_text += text[i-1]
    return new_text

def bintohex(text):
    new_text = ''
    for i in range(16):
        temp = text[i*4:i*4+4]
        temp = int(temp, 2)
        temp = hex(temp)
        temp = temp[2]
        new_text +=  temp
    return new_text
    
```

```python
def DES(mssg, key):
    mssg = textTobit(mssg)
    all_keys = keygenerator(key)
    mssg_IP = permutateIP(mssg)
    for i in range(16):
        l0 = mssg_IP[:32]
        r0 = mssg_IP[32:]
        newl0 = r0
        r0 = convert32_48(r0)
        r0 = xoroperation(r0,all_keys[i])
        r0 = convert48_32(r0)
        r0 = p_each_round(r0)
        r0 = xoroperation(l0,r0)
        mssg_IP = newl0 + r0
    
    r16l16 = mssg_IP[32:] + mssg_IP[:32]
    r16l16 = p_final_round(r16l16)
    
    
    return (bintohex(r16l16).upper())
```
